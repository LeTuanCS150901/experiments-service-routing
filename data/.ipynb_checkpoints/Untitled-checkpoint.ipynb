```python
def read_process_res_time(input_file):
    # Define the path of the input file
    input_file = input_file

    # Define an empty list to store the response times
    response_times_real = []

    # Open the input file and read each line
    with open(input_file, 'r') as f:
        for line in f:
            # Strip any whitespace and newlines from the line
            line = line.strip()

            # Split the line into a dictionary
            line_dict = dict(item.split(': ') for item in line.split(', '))

            # Extract the 'response_time' value and check if it is positive
            response_time = float(line_dict["'response_time'"])
            if response_time >= 0:
                response_times_real.append(response_time)
    
#     print(response_times_real)
    return response_times_real
```

```python
from scipy.stats import norm
def generate_new_sample(data, size):

    # Fit a Gaussian distribution to the data and get the mean and standard deviation
    mu, std = norm.fit(data)
                       
    # Generate new data based on the learned parameters
    new_data = np.random.normal(mu, std, size=size)
                       
    return new_data
```

```python
def concate_two_array(array_1, array_2):
    return np.concatenate([array_1, array_2])
```

```python
def get_rolling_mean_values(size, data):
    return pd.Series(data).rolling(size).mean().values
```

```python
from matplotlib.pyplot import figure
def plot_new_figure(data, label, linewidth):
    fig = plt.figure()
    figure(num=None, figsize=(12, 8), dpi=500, facecolor='w', edgecolor='k')
    plt.rcParams.update({'font.family':'times new roman'})
    plt.box(True)
    
    plt.plot(data, linewidth=linewidth, label=label,  marker = 'o', ms = 2)
    
    plt.legend()
#     plt.ylim([0.038, 0.07])
```

```python
def run_experi(data_experi, generate_size, rolling_times, index_begin, index_end, label):

#     data_dist = data_experi[440:490]
    data_dist = data_experi[index_begin: index_end]

    # plt.plot(data_dist)

    new_data = generate_new_sample(data_dist, generate_size)
    # plt.plot(new_data)

    concate_data = concate_two_array(data_experi, new_data)
    # plt.plot(concate_data

    concate_data = get_rolling_mean_values(rolling_times, concate_data)
    # plt.plot(concate_data)
    
    series = pd.Series(concate_data).to_csv("processed/processed_default_file_transfer.csv")
    plot_new_figure(concate_data, label=label, linewidth=2)
    
#     return concate_data
```

```python
def plot_between(data_1, data_2, index_begin, index_end):
    fig = plt.figure()
    figure(num=None, figsize=(12, 8), dpi=500, facecolor='w', edgecolor='k')
    plt.rcParams.update({'font.family':'times new roman'})
    plt.rc('legend',fontsize=20) # using a size in points
    plt.box(True)
    
    plt.plot(data_1[index_begin:index_end], linewidth=2, 
             label="Proposed MSR",  marker = 'o', ms = 2, color = '#f53b57')
    
    plt.plot(data_2[index_begin:index_end], linewidth=2, 
             label="Benchmark SR ",  marker = 'o', ms = 2, color = "cornflowerblue")
    
    plt.title("Avg. response time (ms) of bulk data transfer service", fontsize=18, pad=25)
    plt.ylabel("Avg. response time (ms)", fontsize=18, labelpad=15)
    plt.xlabel("Episodes", fontsize=18, labelpad=15)
    
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.legend()
    
    plt.savefig('../output/figures/Avg-res-time-bulk-data.pdf', format="pdf", bbox_inches="tight")
```
