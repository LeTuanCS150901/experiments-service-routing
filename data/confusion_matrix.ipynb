```python
import matplotlib.pyplot as plt
import numpy as np

A = [ 
 [1335, 0,0 ,0],
 [0, 942, 0 ,310],
 [0, 1, 702, 0],
 [0, 9,0 ,1154],
]

# Get the confusion matrix
cm = A

# Normalize the confusion matrix by row
cm = cm / np.sum(cm, axis=1, keepdims=True) * 100

# Plot the confusion matrix
fig, ax = plt.subplots(figsize=(12, 8))
plt.imshow(cm, cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.colorbar()

# Set the x-ticklabels
# ax.set_xticklabels(["interactive data", "bulk data transfer", "e-commerce", "video on-demand"])

# Add the labels to the confusion matrix
for i in range(len(cm)):
    for j in range(len(cm[0])):
        plt.text(j, i, "%.2f" % cm[i, j], horizontalalignment="center", verticalalignment="center")

        

# Show the confusion matrix
plt.show()
```

```python
import seaborn as sns


sns.heatmap(cm, annot=True, 
            fmt='.2%', cmap='Blues')
plt.figure(figsize=(12, 12))
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")


bar_width = 0.01

labels = ["interactive data", "bulk data transfer", "e-commerce", "video on-demand"]

# Set the x-tick labels and positions
plt.xticks([r + 0.1 for r in range(4)], labels)


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt  

A = [ 
 [1335, 0,0 ,0],
 [0, 942, 0 ,310],
 [0, 1, 702, 0],
 [0, 9,0 ,1154],
]

# Get the confusion matrix
cm = A

# Normalize the confusion matrix by row
cm = cm / np.sum(cm, axis=1, keepdims=True) * 100
cm = np.around(cm, 2)

labels = ["interactive data", "bulk data transfer", "e-commerce", "video on-demand"]

fig, ax = plt.subplots(figsize=(12, 8), num=None, dpi=500, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.family':'times new roman'})
plt.box(True)
sns.heatmap(cm, annot=True, fmt='g', ax=ax,cmap='Blues', annot_kws={'fontsize':14}) 
#annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
ax.set_xlabel('Predicted labels', fontsize=18, labelpad=15)
ax.set_ylabel('True labels', fontsize=18, labelpad=15)
ax.set_title('Confusion Matrix', fontsize=20, pad=15) 
ax.xaxis.set_ticklabels(labels)
ax.yaxis.set_ticklabels([]);
ax.tick_params(axis='both', which='major', labelsize=16)

plt.savefig('../output/figures/confusion_matrix.pdf', format="pdf", bbox_inches="tight")
```
